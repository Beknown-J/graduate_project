>从上述结果可以看出来，样本内拟合效果确实得到了很大的提升，但是样本外的结果却更差了。造成这个的原因可能主要是因为拟合时pred_Customers特征被训练的参数与预测时参数不同，因为train中的pred_Customers和valid中的pred_Customers存在一定误差，这个误差会造成预测时更大的误差，因此导致结果偏差更大了。这个思想可能是走不通的，除非pred_Customers在样本内和样本外是相同误差。
